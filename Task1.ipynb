```python
import numpy as np
import matplotlib.pyplot as plt
import statistics

months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
income = [4000, 4200, 4500, 4100, 4700, 4300, 4800, 5000, 4700, 4900, 5200, 5100]
expenses = [3000, 3200, 3300, 3100, 3500, 3200, 3400, 3600, 3500, 3700, 3800, 3600]
savings = np.subtract(income, expenses)
```

```python
plt.plot(months, income, label='Income', color='green')
plt.title('Monthly Income')
plt.xlabel('Month')
plt.ylabel('Income')
plt.show()
```

```python
plt.bar(months, savings, color='blue')
plt.title('Monthly Savings')
plt.xlabel('Month')
plt.ylabel('Savings')
plt.show()
```

```python
plt.pie(savings, labels=months, autopct='%1.1f%%', colors=plt.cm.Paired(np.arange(len(months))))
plt.title('Monthly Savings Distribution in Total Yearly Savings')
plt.show()
```

```python
quarters = [income[:3], income[3:6], income[6:9], income[9:12]]
average_income_per_quarter = [statistics.mean(quarter) for quarter in quarters]

print("Average income per quarter:", average_income_per_quarter)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

car_crashes = sns.load_dataset('car_crashes')
```

```python
# Total crashes vs Insurance Premium
plt.figure(figsize=(10, 6))
sns.scatterplot(x='ins_premium', y='total', size='speeding', hue='alcohol', data=car_crashes)
plt.title('Total Crashes vs Insurance Premium (Sized by Speeding %)')
plt.xlabel('Insurance Premium')
plt.ylabel('Total Crashes')
plt.show()
```

```python
# Correlation of car crashes' factors
plt.figure(figsize=(10, 6))
sns.heatmap(car_crashes.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap for Car Crash Factors')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
car_crashes['alcohol'] = car_crashes['alcohol'].round(1)
sns.barplot(x='alcohol', y='total', data=car_crashes)
plt.title('Alcohol-Impaired Crashes by Total Crashes')
plt.xlabel('Alcohol-Impaired (%)')
plt.ylabel('Total Crashes')
plt.xticks(rotation=0) 
plt.show()
```
